```python
# function to scrape data from a givne url
def get_whole_content(url):
    html = urllib.request.urlopen(url).read().decode('utf8')
    text = get_text(html)
    return text
```

```python
faulty_indexes = list()

# iterate 
for index, row in df.iterrows():
    try:
        row['content'] = get_whole_content(row['url'])
        print("index ", index)
    except Exception:
        print(index)
        faulty_indexes.append(index)
```
