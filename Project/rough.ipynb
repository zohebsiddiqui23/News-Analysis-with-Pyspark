```python
import pandas as pd
import ast

# Read the CSV file
df = pd.read_csv("financial_news_complete_content.csv")

# Function to extract name from dictionary string
def extract_name(source_str):
    try:
        # Convert string representation of dict to actual dict
        source_dict = ast.literal_eval(source_str)
        # Extract the 'name' value
        return source_dict.get('name', source_str)
    except:
        # If parsing fails, return the original string
        return source_str

# Apply the function to the source column
df['source'] = df['source'].apply(extract_name)

# Save the updated dataframe back to CSV
df.to_csv("financial_news_complete_content_updated.csv", index=False)

# Display the first few rows to verify
print(df[['source']].head())
```
